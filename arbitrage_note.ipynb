##Porfolio value under the arbitrage free

Fix a portfolio $ h = (x,y) $ , define value process of $ h $ by

$$
V^{h}_{t}=xB_{t}+yS_{t} \text{ , for t=0,1}
$$

More detail ,

$$
\begin{array}{l}
V_{0}^{h}=x+ys\\
V_{1}^{h}=x(1+r)+ysZ
\end{array}\
$$

Definition : an arbitrage portfolio is that a portfolio $ h $ with the following prop.

$$
\begin{array}{l}
V_{0}^{h}=0\\
V_{1}^{h}>0 \text{ , with probability 1}\
\end{array}\
$$

--------------------------------------
Theorom : The model above is free of arbitrage iff <font color='red'>**$d \leq(1+r) \leq u$**</font>\
pf:\
$(\Rightarrow)$\
Suppose on the contrary that $ (1+r)>u $ which implies $ s(1+r)>su $. By assumption before, $ u>d $, then $ s(1+r) > sd $.\
By the two inequalities, we know that either the price of the stock up or down there is always arbitrage if we put money in the bank. Get contradict.\
$(\Leftarrow)$\
Suppose the inequality holds,consider a portfolio $ h $ s.t. $V_{0}^{h}=0$ then we have $ x+ys = 0 $ , which implies that

$$
V_{1}^{h}=
\begin{cases}
ys\ [\ u-(1+r)\ ]\text{ , } \text{if  } Z=u\\
ys\ [\ d-(1+r)\ ]\text{ , } \text{if  } Z=d
\end{cases}
$$

Then, we have two cases, $ y > 0 $ and $ y < 0 $.\
For $ y > 0 $, we have that

$$
\text{$h$ is arbitrage iff}
\begin{cases}
u>1+r\\
d>1+r
\end{cases}
$$

which is contradict. Similarly for $ y < 0 $, then we get that $ h $ is arbitrage free.

------------------
Consequently, we can consider $ 1 + r $ is a convex combination of $ u $ and $ d $, writing

$$
1 + r = q_{u}*u + q_{d}*d
$$

where $ q_{u} $ and $ q_{d} $ can be viewed as a new probability measure $ Q $ with

$$
\begin{cases}
Q(Z=u)=q_{u}\\
Q(Z=d)=q_{d}
\end{cases}
$$

Calculate

$$
\frac{1}{1+r}E^{Q}[S_{1}]=\frac{1}{1+r}[q_{u}su+q_{d}sd]=\frac{1}{1+r}s(1+r)=s
$$

We have

$$
S_{0}=\frac{1}{1+r}E^{Q}[S_{1}]
$$

, called risk neutral valuation formula, such probability measure $ Q $ is called a Martingale measure.

Motivation : We want to build a connection between prices, based on different times, by such probability measure, we call it Martingale measure.

Proposion : The market model is arbitrage free iff there exists a martingale measure $ Q $

Proposion : The martingale probabilities are

$$
\begin{cases}
q_{u}=\frac{(1+r)-d}{u-d}\\
q_{d}=\frac{u-(1+r)}{u-d}
\end{cases}
$$





## Reasonable contigent claim price

Definition : contingent claim is a derivative whose future payoff depends on the value of another “underlying” asset, or more generally, that is dependent on the realization of some uncertain future event.

In this chapter ,contingent Claim viewed as Option.\
Denote the price of a given contingent claim $ X $ at time $ t $ by $ \Pi(t;X) $

$$
\text{Our purpose is to seek $ \Pi(0;X) $ such that $ \Pi(1;X) = X $}
$$

----------------
Definition : The function $ \Phi $ is called the <font color='red'>**contract function**</font> which is to describe the net profit of some contingent claim at the every moment.\
For example : A European call option with strike price $ K $, we assume that $ sd < K < su $. Acturally, we can still get the contract function without this assumption. Namely, $ K < sd $ and $ K > su $ are allowed. We pay $ K $ to get the stock and sell if $ S_{1} > K $, so we will have a net value of $ su - K $.\
Hence, the contract function given by

$$
\begin{cases}
Φ(u) = su-K\\
Φ(d) = 0
\end{cases}
$$

-----------
Definition : A contingent claim $ X $ is said to be reachable if there exists a portfolio $ h $, called a hedging or replicating portfolio, such that

$$
V^{h}_{1}=X
$$

If all contingent claims can be replicated we say the market is complete.

------------
Pricing Principle : If a claim $ X $ is reachable with replicating portfolio $ h $, then the only reasonable price process for $ X $ is given by

$$
\Pi(t;X) = V^{h}_{t} , t = 0,1
$$

Reasonalbe means arbitrage free.

-------------
Proposion : If the general binomial model is arbitrage free, then it is complete.

pf :\
First, fix an arbitrage claim $X$ with contract function $ \Phi $, we want to show that there exists a portfolio $ h = (x,y) $ s.t.

$$
\begin{cases}
V_{1}^{h} = Φ(u) \text{ , if } Z = u\\
V_{1}^{h} = Φ(d) \text{ , if } Z = d
\end{cases}
$$

which implies that

$$
\begin{cases}
(1+r)x + suy = \Phi(u),\\
(1+r)x + sdy = \Phi(d).
\end{cases}
$$

By assumption $ u > d $, the above linear system has a unique solution given by

$$
\begin{cases}
x=\frac{1}{1+r} \frac{u\Phi(d)-d\Phi(u)}{u-d},\\
y=\frac{1}{s} \frac{\Phi(u)-\Phi(d)}{u-d}.
\end{cases}
$$

-------------
Proposition : If the binomial model is arbitrage free, then the arbitrage free price of a contingent claim $ X $ is given by

$$
\Pi(0;X) = \frac{1}{1+r}E^{Q}[X]
$$

pf :

$$
\Pi(0;X) = x + sy = \frac{1}{1+r}(\frac{(1+r)-d}{u-d}*Φ(u)+\frac{u-(1+r)}{u-d}*Φ(d))
$$

## Example


We set $ S_{0} = 100, u = 1.2, d = 0.8, p_{u} = 0.6, p_{d} = 0.4, r = 0$\
Thus we have the price dynamics

$$
\begin{array}{l}
S_{0}=100,\\
S_{1}=
\begin{cases}
120 \text{ , with probility 0.6}\\
80 \text{ , with probility 0.4}
\end{cases}
\end{array}
$$

Tomorrow's price we get

$$
\frac{1}{1+r}E^{P}[S_{1}] = 1*[120*0.6+80*0.4] = 104
$$

This is higher than the value of today's stock price of 100.

We consider a European call with strike price $ K = 110 $ , so the claim $ X $ is given by

$$
X =
\begin{cases}
\text{$10$ , if $S_{1} = 120$}\\
\text{ $0$ , if $S_{1} = 80$}
\end{cases}
$$

Then, some people might calculate the price as the discounted value using objective probabilities, resulting in a price of

$$
\Pi(0;X) = \frac{1}{1+0} * [10*0.6 + 0*0.4] = 6
$$

But if we use the theory above, it is easily seen that the martingale probabilities are given by $ q_{u} = q_{d} = 0.5 $ ,thus giving us the theoretical price

$$
\Pi(0;X) = \frac{1}{1+0} * [10*0.5 + 0*0.5] = 5
$$

The theoretical price differ from the discounted value using objective probabilities.\
So, we want to check the theory is correct or not.\
First, we should be able to replicate the option. It menans we can find a portfolio $ h $ with fit $ x $ and $ y $ such that $ V_{1}^{h} = X $

By theorem, we know that

$$
\begin{array}{l}
x = \frac{1.2*0-0.8*10}{1.2-0.8} = -20\\
y = \frac{1}{100}*\frac{10-0}{1.2-0.8} = \frac{1}{4}
\end{array}
$$

this means that the replicating porfolio is formed by borrowing $ \$20 $ from the bank, and investing this money in a quarter of a share in the stock.

Thus,the value of the portfolio at $ t = 1 $ is

$$
\begin{array}{l}
V_{1}^{h} = -20 + \frac{1}{4}*120 = 20 \text{ , if $S_{1} = 120$}\\
V_{1}^{h} = -20 + \frac{1}{4}*80 =  0 \text{ , if $S_{1} = 80$}
\end{array}
$$

this value is same as the $ X $

And we find that when time $ t = 0 $, the portfolio value is $ \$5 $ ; therefore, we define the option price as $ \$5 $

This may sound reasonable, because someone who sells the option get $ \$5 $ dollars, and he borrows $ \$20 $ dollars from the bank , then he has $ \$25 $ dollars just enough to buy a quarter of a share which price is $ 100*\frac{1}{4} = 25 $ in the stock .

So, if someone sells the option at a price different from $ \$5 $, we can buy it and obtain a risk-free arbitrage.

## T-claim price with multiperiod model

Consider a T-claim $ X = Φ(S_{T}) $. Then this claim can be replicated using a self-financing portfolio. If $ V_{t}(k) $ denotes the value of the portfolio at the node $ (t,k) $ then $ V_{t}(k) $ can be computed recursively

$$
\begin{cases}
V_{t}(k) = \frac{1}{1+r} \text{ \{ } q_{u}V_{t+1}(k+1) + q_{d}V_{t+1}(k)\text{ \} }\\
V_{t}(k) = \Phi(\ su^{\ k}d^{\ (t-k)}\ )
\end{cases}
$$

where the martingale probabiities $ q_{u} $ and $ q_{d} $ are given by

$$
\begin{cases}
q_{u} = \frac{(1+r)-d}{u-d}\\
q_{d} = \frac{u-(1+r)}{u-d}
\end{cases}
$$

With the notation as above, the hedging portfolio is given by

$$
\begin{cases}
x_{t}(k) = \frac{1}{1+r} * \frac{uV_{t}(k) - dV_{t}(k+1)}{u-d}\\
y_{t}(k)= \frac{1}{S_{t-1} } * \frac{ V_{t}(k+1) - V_{t}(k) }{u-d}
\end{cases}
$$

Then we can know the arbitrage free price at $ t = 0 $ of a T-claim $ X $ is given by

$$
\Pi(0;X) = \frac{1}{(1+r)^{T}} * E^{Q}[X]
$$

where $ Q $ denotes the martingale measure, or more explicitly

$$
\Pi(0;X) = \frac{1}{(1+r)^{T}} \sum_{k=0}^{T} \binom{T}{K}q_{u}^{k}q_{d}^{T-k}\Phi(\ su^{\ k}d^{\ (T-k)}\ )
$$

The condition $ d < (1+r) < u $ is a necessary and sufficient condition for absence of arbitrage.

# 3. The General Model

## Model

We consider $N$ different assets here and only two points in time, $ t = 0 $ and $ t = 1 $\
Hence, we have such price vector process $ S_{t}, t = 0, 1 $

$$
S_t=
\begin{bmatrix}
S_{t}^{1}\\
.\\
.\\
.\\
S_{t}^{N}
\end{bmatrix}
$$

--------
Assumpt that we have finite sample space $\ Ω = {ω_{\ 1}, ..., ω_{\ M}} $ and the probability $ p_{j} = P(ω_{\ j}),\ j = 1,...,N $ are all strictly positive.

We define the matrix $ D $ by

$$
D=
\begin{bmatrix}
S_{1}^{1}(ω_{\ 1})&\ S_{1}^{1}(ω_{\ 2})&\ ...&\ S_{1}^{1}(ω_{\ M})\\
S_{1}^{2}(ω_{\ 1})&\ S_{1}^{2}(ω_{\ 2})&\ ...&\ S_{1}^{2}(ω_{\ M})\\
.&\ .&\ &. \\
.&\ .&\ &. \\
.&\ .&\ &. \\
S_{1}^{N}(ω_{\ 1})&\ S_{1}^{N}(ω_{\ 2})&\ ...&\ S_{1}^{N}(ω_{\ M})\\
\end{bmatrix}
$$

We also write $D$ as

$$
D = S_{t} =
\begin{bmatrix}
d_{1}...d_{M}
\end{bmatrix}
$$

where $ d_{1}, ..., d_{M} $ are the columns of $ D $.

-------
Then, we have the following assumptions

$$
S_{0}^{1} > 0,\\
S_{1}^{1}(ω_{\ j}) > 0, j = 1,...,M.
$$

-------
Now we define a portfolio with an $ N $ dimensinal row vector weight $ h = [h^{1},...,h^{N}] $ , then the value process of our portfolio will be a stochastic process $ V_{t}^{h} $ defined by

$$
V_{t}^{h} = \sum_{i=1}^{N}h^{i}S_{t}^{i}=hS_{t},\ \  t=0,1
$$

we can write more detail

$$
V_{t}^{h}(w_{i}) = hS_{t}(w_{i}) = hd_{i} = (hD)_{i}
$$

---
Definition : The portfolio $ h $ is an arbitrage portfolio if it satisfies the conditions

$$
\begin{array}{r}
V_{0}^{h}=0\\
P(V_{1}^{h}\ge0)=1\\
P(V_{1}^{h}>0)>0
\end{array}
$$

we can write more detail

$$
\begin{array}{r}
V_{0}^{h} = 0 &\\
V_{1}^{h}(w_{i})\geq 0 &\text{, for all } i = 1,...,M \ \ \ \ \\
V_{1}^{h}(w_{i}) > 0 &\text{, for some } i = 1,...,M
\end{array}
$$

We now go on to investigate when the market model above is free of arbitarge possibilities, and the main technical tool for this investigation is the Farkas' Lemma.

## Farkas' Lemma

Suppose that $ d_{0}, d_{1}, ..., d_{K} $ are column vectors in $ R^{N} $.
Then, exactly one of the two following problems possesses a solution.\
Problem1 : Find nonnegative number $ λ_{1}, ..., λ_{K} $ such that

$$
d_{0}=∑\limits_{j=1}^{K}λ_{j}d_{j}.
$$

Problem2 : Find a row vector $h\in R^{N}$ such that

$$
\begin{array}{l}
h*d_{0} < 0\\
h*d_{j} \geq 0, j=1, ..., M
\end{array}
$$

pf :

Let $ K $ be the set of all nonnegative linear combinations of $ d_{1}, ..., d_{M} $\
Then, one of the two cases is hold.
* The vector $ d_{0} \in K$, which implie that the Problem1 has a solution.
* The vector $ d_{0} \notin K$, then there exists a hyperplane $ H $ such that $ d_{0} $ is strictly on one side of $ H $ and $ K $ is on the other side.

Letting $ h $ be a normal vector to $ H $ pointing in the direction where $ K $ lies, which implies that the Problem2 has a solution.




## Normalized price system

We go on to investigate absence of arbitrage for our model.\
In order to see more clearly what is going on we will consider not only the nominal price system $ S^{1},...,S^{N} $,but also the normalized price system $ Z^{1},...,Z^{N} $, under the numeraire $ S^{1} $, defined by

$$
Z_{t}^{i}(w_{j}) = \frac{S_{t}^{i}(w_{j})}{S_{t}^{1}(w_{j})},\ \  i = 1,...,N , \ \ j = 1,...,M.
$$

It helps us to easy translation between two price systems.

----
Now, we note that a given portfolio $ h $ can be viewed as a portfolio in $ S $ system and as a porfolio in the $ Z $ system.

$$
V_{t}^{h} = hS_{t} \text{ for $ S $ system}\\
V_{t}^{h,Z} = hZ_{t} \text{ for $ Z $ system}
$$

With notation as above, the following hold.

1. The $ Z $ value process is related to $ S $ value process by

$$
V_{t}^{h,Z} = \frac{1}{S_{t}^{1}}V_{t}^{h}
$$

2. A porfolio is an arbitrage in $ S $ economy $ \Leftrightarrow $ it is an arbitrage in the $Z$ economy.

3. In the $ Z $ price system, the numeraire asset $ Z^{1} $ has unit constant price, i.e.

$$
Z_{0}^{1} = 1\\
Z_{1}^{1}(w_{j}) = 1, \ \ j = 1,...,M.
$$

We can find that $ Z^{1} ≡ 1 $ is that in the normalized price system, the numeraire asset is risk free corresponding to a bank with zero interest rate.

-----
Theorem(***) : The market is arbitrage free $ \Leftrightarrow $ there exists strictly positive real numbers $ q_{1},...,q_{M} $ with

$$
q_{1} + ... + q_{M} = 1,
$$

such that the following vector equality holds

$$
Z_{0} = \sum_{j=1}^{M}Z_{1}(w_{j})q_{j},
$$

or,on component form,

$$
Z_{0}^{i} = \sum_{j=1}^{M}Z_{1}^{i}(w_{j})q_{j}, \ \ i = 1,...,N.
$$

pf :
Define the matrix $ D^{Z} $ by

$$
D^{Z} =
\begin{bmatrix}
Z_{1}^{1}(ω_{\ 1})&\ Z_{1}^{1}(ω_{\ 2})&\ ...&\ Z_{1}^{1}(ω_{\ M})\\
Z_{1}^{2}(ω_{\ 1})&\ Z_{1}^{2}(ω_{\ 2})&\ ...&\ Z_{1}^{2}(ω_{\ M})\\
.&\ .&\ &. \\
.&\ .&\ &. \\
.&\ .&\ &. \\
Z_{1}^{N}(ω_{\ 1})&\ Z_{1}^{N}(ω_{\ 2})&\ ...&\ Z_{1}^{N}(ω_{\ M})\\
\end{bmatrix}
$$

which we can also write as

$$
D^{Z} =
\begin{bmatrix}
1&\ 1&\ ...&\ 1\\
Z_{1}^{2}(ω_{\ 1})&\ Z_{1}^{2}(ω_{\ 2})&\ ...&\ Z_{1}^{2}(ω_{\ M})\\
.&\ .&\ &. \\
.&\ .&\ &. \\
.&\ .&\ &. \\
Z_{1}^{N}(ω_{\ 1})&\ Z_{1}^{N}(ω_{\ 2})&\ ...&\ Z_{1}^{N}(ω_{\ M})\\
\end{bmatrix}
$$

It is clear that the market is arbitrage free if and only if the following system of equations has no solution

$$
\begin{array}{r}
hZ_{0} = 0 , \\
(hD^{Z})_{j} \geq 0, &\text{for all } j = 1,...,M \ \ \ \\
(hD^{Z})_{j} > 0, &\text{ for some } j = 1,...,M
\end{array}\\
$$
$\text{ where } h \in R^{N}$

Now we apply Farkas' Lemma to this system and in order to do this we define the column vector $ p $ in $ R^{M} $ by

$$
p =
\begin{bmatrix}
p_{1}\\
p_{2}\\
.\\
.\\
.\\
p_{M}\\
\end{bmatrix}
$$

we can now rewrite the system above as

$$
\begin{align}
hZ_{0} = 0,\\
hD^{Z} \geq 0,\\
hD^{Z}p > 0,
\end{align}
$$

we rewrite the first equality as a double inequality to obtain

$$
\begin{align}
hZ_{0} \geq 0,\\
h(-Z_{0}) \geq 0,\\
hD^{Z} \geq 0,\\
-hD^{Z}p < 0,
\end{align}
$$

The point of all this is that if we define the vector $ \hat{{d}_{0}} $ and the block matrix $ \hat{D} $ by

$$
\begin{align}
\hat{{d}_{0}} &= -D^{Z}p,\\
\hat{D} &= [D^{Z}, Z_{0} ,-Z_{0}]
\end{align}
$$

we see that the market is free of arbitrage if and only if the system

$$
h\hat{{d}_{0}} < 0,\\
h\hat{D} \geq 0,
$$

has no solution $ h \in R^{N} $

Applying Farkas' Lemma to this system we thus see that absencs of arbitrage is equivalent to the existence of non negative real numbers $\lambda_{1},\lambda_{2},...,\lambda_{M},\lambda_{N+1},\lambda_{N+2} $ such that

$$
\hat{d}_{0} = \hat{D}\lambda
$$

If we define the vector $\beta \in R^{M} $ by

$$
\beta =
\begin{bmatrix}
\lambda_{1}\\
\lambda_{2}\\
.\\
.\\
.\\
\lambda_{M}
\end{bmatrix}
$$

and define the real number $ \alpha = λ_{N+2} - \lambda_{N+1} $ we can write the relation $\hat{d}_{0} = \hat{D}\lambda $ as

$$
-D^{Z}p = D^{Z}\beta - \alpha Z_{0}
$$

or

$$
\alpha Z_{0} = D^{Z}(p+\beta)
$$

Here $\beta \geq 0 $ but we don't know the sign of $ \alpha $

Recall that $ Z_{0}^{1} = 1 $ and the first row of $ D^{Z} $ has the unit 1 in all positions, we obtain

$$
\alpha = \sum_{j=1}^{M}(p_{j}+\beta_{j})
$$

From this we see that $\alpha > 0 $, and if we define the column vector $ q \in R^{M} $ by

$$
q = \frac{1}{\alpha}(p+\beta)
$$

we see that

$$
q_{j} > 0, j = 1,...,M,
$$

and that

$$
\sum_{i=1}^{M}q_{j} = 1
$$

we get equation

$$
Z_{0} = D^{Z}q
$$

## Martingale measures

Definition : Given the objective probability measure $ P $, we say that another probability measure $ Q $ defined on the sample $ Ω $ is equivalent to $ P $ if

$$
P(A) = 0 \Leftrightarrow Q(A) = 0
$$

or

$$
P(A) = 1 \Leftrightarrow Q(A) = 1
$$

Definition : A discrete time random process $ \lbrace \ X_{n}: n = 0,1,...\rbrace $ is a martingale if

$$
E\ \lbrack X_{m}|𝑭_{n}\rbrack = X_{n}, \text{for all } n \leq m
$$

where $ 𝑭_{n} $ denotes the information available at time $ n $.

Definition : Consider the market model we set in this chapter, and fix the asset $ S^{1} $ as the numeraire asset. We say that a probability measure $ Q $ defined on $ Ω $ is a martingale measure if it satisfies the followings.


1. $ Q $ is equivalent to  P $
2. For every $ i = 1, ..., N $, the normalized asset price process

$$
Z_{t}^{i} = \frac{S_{t}^{i}}{S_{t}^{1}}
$$

is a martingale under the measure $ Q $.


## First Fundamental Theorem

Theorom : Given a fixed numeraire, the market is free of arbitrage possibilities if and only if there exists a martingale measure $ Q $.

proof :
From above theorem(***) we know that absence of arbitrage is equivalent to the existence of strictly positive constant $ q_{1},...,q_{M} $ with $q_{1} + ... + q_{M} = 1 $, we may thus define a probability measure $ Q $ by setting $Q(w_{j}) = q_{j} $ and since $q_{j} > 0 $ for all $ j $ we see that $ Q ∼ P $. With this definition of $ Q $, we have

$$
Z_{0} = E^{Q}[Z_{1}]
$$

----
Theorom : Assume that there exist a risk free asset, and denote the corresponding risk free interest rate bt $ R $. Then the market is arbitrage free if and only if there exist a measure $ Q ∼ P $ such that

$$
S_{0}^{i} = \frac{1}{1+r}E^{Q}[S_{1}^{i}], \text{ for all } i = 1,...,N.
$$

pf :
Under the martingale condition above, we assume that the numeraire asset $. S^{1} $ is a risk free asset, in the sense that the asset price $ S_{1}^{1} $ at time $ t = 1 $ is constant as a function of $ w $.
We assume $ S_{0}^{1} = 1 $, so we get

$$
S_{1}^{1}(w_{j}) = 1 + R, \text{ for all } j = 1,...,M,
$$

then we discounted with the risk free rate, we can get

$$
S_{0}^{i} = \frac{1}{1+r}E^{Q}[S_{1}^{i}], \text{ for all } i = 1,...,N.
$$

----
Note :
the expectation above is not taken under the objective probability measure $ P $ but under the martingale measure $ Q $.

## Complete market

We assume that the market $ S^{1}, S^{2}, ..., S^{N} $ is arbitrage free and there exists a risk free asset.

----
Definiton : Consider a contingent claim $ X $. If there exists a portfolio $ h $, based on the underlying assets such that

$$
V_{1}^{h}=X \text{, with probability 1}\\
\text{i.e. } V_{1}^{h}(ω_{j}) = X(ω_{j}), j=1,...,M
$$

then we say that $ X $ is replicated or hedged by $ h $\
Such $ h $ is called a replicationg or hedging portfolio. If every contingent claim can be replicated, we say that the market is complete.

----
Theorom : The market is complete iff the rows of the matrix $D$ span $R^{M}$

pf :
For any portfolio $h$, we view the random variable $ V_{1}^{h} $ as a row vector $ [V_{1}^{h}(ω_{\ 1}), ..., V_{1}^{h}(ω_{\ M})] $ then we have

$$
V_{1}^{h}=hD
$$

Thus, the market is complete iff for every random variable $ X $, the equation

$$
hD=X
$$

has a solution. \
And we know $hD$ is exactly a linear combination of the rows of $ D $.
Done.

----
Therefore, assume that the claim $ X $ can be replicated by the portfolio $ h $, then there is an obvious candidate as the price at time $ t=0 $ for $ X $.

$$
Π(0;X) = V_{0}^{h}
$$

which is the natural pricing formula.

We note that any other price would produce an arbitrage possibility.\
For example $ V_{0}^{h} = 7 $ and $ Π(0;X) = 10 $, then we would obtain 3 units by selling the claim and buing the replicated portfolio.

The pricing formula can also be written as

$$
Π(0;X)=\frac{1}{1+r}E^{Q}[X]
$$
\
Since when we assume $h$ replicates $X$, then by definition
$$
X=hS_{1}
$$
Then, we get
$$
Π(0;X)=hS_{0}
$$
And since on arbitrage free market we have such the pricing formula
$$
S_{0}=\frac{1}{1+r}E^{Q}[S_{1}]
$$
Combining the two equations above, we get this pricing formula.

## Second Fundamental Theorem

Theorom : Assume that the model is arbitrage free. Then the market is complete iff the martingale measure is unique.

pf : By Theorom above, we know that the market is arbitrage free iff the rows of $ D $ span $ R^{M} $ i.e.

$$
Im \ [ \ D^{T}] = R^{M}
$$

where $ D^{T} $ is the transpose matrix as a mapping from $ R^{N} $ to $ R^{M} $

From Theorem in absence of arbitrage, we know that there exists a solution to the equation

$$
Z_{0} = D^{Z}q
$$

This solution is unique iff $ Ker\ [\ D^{Z}] = {0} $, which is equivalent to $ Ker \ [D] = {0} $

Then, by duality result

$$
(Im \ [ \ D^{T}])^{⊥} = Ker[D]
$$
\
Thus,

$$
Ker \ [D] = 0 \Leftrightarrow Im \ [ \ D^{T}] = R^{M}
$$

i.e. the martingale measure is unique iff the market is complete . Done.
